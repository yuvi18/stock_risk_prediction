```python
import numpy
from pandas import read_csv
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from gensim.models import Word2Vec
from gensim.models.keyedvectors import KeyedVectors
```

# load dataset

```python
emissions = read_csv('../notebooks/content/emmissions.csv')
emissions.head()
```

```python
columns = ['company', 'country', 'reporting_year', 'symbol', 'isin', 'disclosure_score', 'perfomance_band', 'parent_account', 'permission', 'response_status', 'scope1_co1', 'scope2_co2']
emissions.columns = columns
```

```python
emissions.head()
```

```python
emissions.isnull().sum(axis = 0)
```

```python
emissions = emissions[['company','reporting_year', 'symbol', 'disclosure_score', 'scope1_co1', 'scope2_co2']]
```

```python
emissions.fillna(value={'disclosure_score': 0.0, 'scope1_co1': 0.0, 'scope2_co2': 0.0}, inplace=True)
```

```python
emissions.isnull().sum(axis = 0)
```

```python
emissions.head()
```

```python
emissions['symbol'] = emissions['symbol'].map(lambda x:x.rstrip('US'))
```

```python
emissions.head()
```

```python
binary_vectors = 'stoack2vec_Keyed_Binary.bin'
text_vectors = 'stoack2vec_Keyed_Text.vec'

model = KeyedVectors.load_word2vec_format(text_vectors, binary=False)
word_dict = {}
embeddings = []
symbols = []
for word in model.vocab:
    word_dict[word] = model[word]
    #embeddings.append(model2[word])
    symbols.append(word)
for key in word_dict:
    embeddings.append(word_dict[key])
#embeddings
len(word_dict)
```

```python
stock_symbols = symbols
emissions_symbols = emissions['symbol'].tolist()
```

```python
flag = 0
if(all(symbol in stock_symbols for symbol in emissions_symbols)): 
    flag = 1
# printing result 
if (flag) : 
    print ("Yes, list is subset of other.") 
else : 
    print ("No, list is not subset of other.") 
```
